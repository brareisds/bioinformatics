### Seleção dos genes

Os genes foram escolhidos a partir do artigo de revisão **Verma et al. (2024)** - *"Developments in extremophilic bacterial genomics: A post next generation sequencing era"*, publicado em *Ecological Genetics and Genomics* 32 (2024) 100255.

Os trechos relevantes são:

**dnaK** (chaperona molecular Hsp70) - adaptação ao frio, seção 3.2:
> *"Maillot et al. (2019) reported the role of DnaK (Hsp70) proteins with a dedicated J-domain protein (JDP) in allowing S. oneidensis to survive in a cold environment"*

Função: proteína chaperona que auxilia no dobramento correto de outras proteínas. Em organismos do frio, atua reparando e estabilizando proteínas que poderiam agregar em baixas temperaturas.


**groEL** (chaperonina) - adaptação ao calor, seção 4.2:
> *"HGT event could be one of the reasons for the upregulation of genes encoding heat shock proteins, such as the dnaJ-like genes HtpG, Hsp33, GroEL1/EL2/ES, and HspA"*

Função: chaperonina que auxilia no dobramento de proteínas. Em organismos do calor, é superexpressa para lidar com proteínas desnaturadas pelo calor.


**gyrA** (DNA girase, subunidade A) - adaptação ao frio, seção 3.4:
> *"DNA topoisomerase IV (gyrA and gyrB) [...] play a role in cold adaptation"*

Função: enzima que regula o superenrolamento do DNA. Em organismos do frio, mantém o DNA mais relaxado para permitir transcrição e replicação em baixas temperaturas.


**deaD** (RNA helicase DEAD-box) - adaptação ao frio, seção 3.4:
> *"cold-shock DEAD-box helicase A (deaD) [...] play a role in cold adaptation"*

Função: helicase que desestabiliza estruturas secundárias de RNA que se formam em excesso no frio, garantindo que o RNA permaneça acessível para tradução.


**reverse gyrase** - identificada no KEGG (K03170) como contraparte funcional do gyrA em hipertermófilos. Diferentemente do gyrA, que relaxa o DNA, a reverse gyrase **superenrola positivamente** o DNA, estabilizando-o em altas temperaturas.

### Observações sobre o padrão de presença/ausência

Mesmo antes do alinhamento, o mapeamento revela padrões biologicamente relevantes:

**Reverse gyrase restrita ao calor extremo** - presente em 3 dos 4 hipertermófilos e ausente em todos os mesófilos e psicrófílos. Isso é consistente com sua função de estabilizar o DNA em altas temperaturas - uma necessidade exclusiva de organismos hipertermófilos.

**gyrA ausente nas arqueias hipertermófilas** - *P. furiosus* e *T. litoralis* não possuem gyrA mas possuem reverse gyrase, sugerindo uma substituição funcional associada ao ambiente: o gene de topologia do DNA é trocado por sua versão adaptada ao calor.

**M. burtonii comporta-se como bactéria** - esta archaea psicrófila possui dnaK e gyrA, genes tipicamente bacterianos. Isso é consistente com o fenômeno de **transferência horizontal de genes (HGT)** discutido em Verma et al. (2024), pelo qual organismos de ambientes extremos adquirem genes de outros domínios quando isso confere vantagem adaptativa.

**groEL restrito a Bacteria** - nenhuma archaea possui groEL no conjunto estudado. As arqueias utilizam o **thermosome** como chaperona equivalente - uma proteína funcionalmente análoga mas estruturalmente distinta, que não é detectada pelo KO do groEL.

### 2. Alinhamento das sequências 

#### Download das sequências proteicas

Com os accessions identificados no KEGG, as sequências
proteicas de cada gene para cada organismo são baixadas. Isso é feito diretamente do **NCBI Protein**
usando a biblioteca Biopython.

As sequências são salvas
no formato **FASTA**, um formato de texto simples onde cada sequência é precedida por uma
linha de cabeçalho iniciada com `>` seguida do nome do organismo.

```python
from Bio import Entrez
import os

Entrez.email = "brareisds@gmail.com"

os.makedirs("genes_referencia", exist_ok=True)

genes_por_arquivo = {
    "gyrA":      ["T_ruber", "T_maritima", "E_fergusonii", "P_arcticus", "M_burtonii"],
    "revgyrase": ["T_maritima", "P_furiosus", "T_litoralis"],
    "dnaK":      ["T_ruber", "T_maritima", "E_fergusonii", "P_arcticus", "M_burtonii"],
    "groEL":     ["T_ruber", "T_maritima", "E_fergusonii", "P_arcticus"],
    "deaD":      ["T_ruber", "T_litoralis", "E_fergusonii", "M_maripaludis"],
}

# Para cada gene, baixa as sequências de todos os organismos
# e salva em um único arquivo FASTA
for gene, organismos in genes_por_arquivo.items():
    output = f"genes_referencia/{gene}_todos.faa"
    with open(output, "w") as out:
        for org in organismos:

            # Monta a chave para buscar o accession no dicionário
            chave = f"{org}_{gene}"
            accession = todos_accessions[chave]

            # Faz a requisição ao NCBI para baixar a sequência em formato FASTA
            handle = Entrez.efetch(
                db="protein",       # banco de proteínas do NCBI
                id=accession,       # accession da sequência
                rettype="fasta",    # formato de saída
                retmode="text"
            )

            seq = handle.read().strip().split("\n")
            seq[0] = f">{org}"
            out.write("\n".join(seq) + "\n\n")
            print(f" {gene} | {org} ({accession})")
    print(f" Salvo: {output}\n")
```

### Alinhamento múltiplo das sequências

Com as sequências baixadas, é feito o alinhado, ou seja, posicionar os
aminoácidos equivalentes de cada organismo lado a lado para que seja possível comparar
posição por posição.

O alinhamento é feito com o **MAFFT** (Multiple Alignment using Fast Fourier Transform).



O MAFFT com a opção `--auto` escolhe automaticamente o melhor algoritmo para o conjunto
de sequências. Para todos os 5 genes deste trabalho, o log de execução reportou:

> Strategy: L-INS-i (Probably most accurate, very slow)  
> Iterative refinement method (<16) with LOCAL pairwise alignment information  
> alg=L, model=BLOSUM62, 2.00, -0.10, +0.10

O algoritmo **L-INS-i** calcula alinhamentos locais par a par, usando o mesmo
princípio do algoritmo Smith-Waterman, e os
combina progressivamente para construir o alinhamento múltiplo. 

A matriz de substituição utilizada é a **BLOSUM62**, uma matriz construída a partir de
blocos de sequências proteicas com pelo menos 62% de identidade entre si, que atribui
pontuações a cada par de aminoácidos com base na frequência com que substituições ocorrem
naturalmente na evolução. Pares de aminoácidos quimicamente similares recebem pontuações
positivas; pares muito diferentes recebem pontuações negativas.

O MAFFT recebe como entrada um arquivo FASTA contendo as sequências proteicas
do mesmo gene em organismos de ambientes diferentes. Para cada gene, são alinhadas
as sequências dos organismos que o possuem conforme identificado no mapeamento do KEGG:

| Gene | Sequências alinhadas |
|---|---|
| gyrA | T. ruber, T. maritima, E. fergusonii, P. arcticus, M. burtonii |
| revgyrase | T. maritima, P. furiosus, T. litoralis |
| dnaK | T. ruber, T. maritima, E. fergusonii, P. arcticus, M. burtonii |
| groEL | T. ruber, T. maritima, E. fergusonii, P. arcticus |
| deaD | T. ruber, T. litoralis, E. fergusonii, M. maripaludis |

O resultado é um arquivo FASTA alinhado onde cada coluna representa uma posição
equivalente na proteína, permitindo comparar aminoácido por aminoácido entre
organismos de ambientes térmicos distintos.


```python
import subprocess
import os

# Cria a pasta onde os alinhamentos serão salvos, se ainda não existir
os.makedirs("alinhamentos", exist_ok=True)

# Lista dos genes a serem alinhados
# Cada gene tem um arquivo FASTA com as sequências de todos os organismos que o possuem
genes = ["gyrA", "revgyrase", "dnaK", "groEL", "deaD"]

for gene in genes:
    # Caminho do arquivo de entrada — sequências do gene para todos os organismos
    entrada = f"genes_referencia/{gene}_todos.faa"
    
    # Caminho do arquivo de saída — sequências já alinhadas
    saida = f"alinhamentos/{gene}_alinhado.fasta"

    # Executa o MAFFT via linha de comando
    # --auto: o MAFFT escolhe automaticamente o melhor algoritmo
    # capture_output=True: captura tanto o alinhamento (stdout) quanto o log (stderr)
    # text=True: retorna o resultado como texto em vez de bytes
    resultado = subprocess.run(
        f"mafft --auto {entrada}",
        shell=True,
        capture_output=True,
        text=True
    )

    # O alinhamento em formato FASTA vem no stdout 
    with open(saida, "w") as f:
        f.write(resultado.stdout)

    # O log do MAFFT vem no stderr 
    linhas_relevantes = [
        linha for linha in resultado.stderr.split("\n")
        if any(x in linha for x in ["Strategy", "L-INS", "model=", "alg=L"])
    ]

    # Imprime o resumo do alinhamento para cada gene
    print(f"=== {gene} ===")
    for linha in linhas_relevantes:
        print(" ", linha)
    print()
```

```python
from Bio import SeqIO

# Todas as sequências de um mesmo gene devem ter o mesmo comprimento
# após o alinhamento, caso contrário, gaps foram inseridos para compensar
# diferenças de tamanho entre as sequências.
for gene in genes:
    print(f"\n=== {gene} ===")
    for rec in SeqIO.parse(f"alinhamentos/{gene}_alinhado.fasta", "fasta"):
        print(f"  {rec.id:<20} {len(rec.seq)} aa")
```

```python
from Bio import AlignIO
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

# Configurações
ambientes = {
    "T_ruber":       "hipertermofilo",
    "T_maritima":    "hipertermofilo",
    "P_furiosus":    "hipertermofilo",
    "T_litoralis":   "hipertermofilo",
    "E_fergusonii":  "mesofilo",
    "M_maripaludis": "mesofilo",
    "P_arcticus":    "psicrófilo",
    "M_burtonii":    "psicrófilo",
}


# Classificação química baseada nas propriedades das cadeias laterais dos aminoácidos
# conforme descrito em Lehninger Principles of Biochemistry.
quimica = {
    'A': 'apolar', 'V': 'apolar', 'I': 'apolar', 'L': 'apolar',
    'M': 'apolar', 'F': 'apolar', 'W': 'apolar', 'P': 'apolar',
    'G': 'apolar',
    'S': 'polar',  'T': 'polar',  'C': 'polar',  'Y': 'polar',
    'N': 'polar',  'Q': 'polar',
    'K': 'carregado+', 'R': 'carregado+', 'H': 'carregado+',
    'D': 'carregado-', 'E': 'carregado-',
    '-': 'gap'
}

genes = {
    "gyrA":      ["T_ruber", "T_maritima", "E_fergusonii", "P_arcticus", "M_burtonii"],
    "dnaK":      ["T_ruber", "T_maritima", "E_fergusonii", "P_arcticus", "M_burtonii"],
    "groEL":     ["T_ruber", "T_maritima", "E_fergusonii", "P_arcticus"],
    "deaD":      ["T_ruber", "T_litoralis", "E_fergusonii", "M_maripaludis"],
}

resultados = {}

for gene, organismos in genes.items():
    
    # Carrega o alinhamento do gene e cria um dicionário {organismo: sequência}
    aln = AlignIO.read(f"alinhamentos/{gene}_alinhado.fasta", "fasta")
    seqs = {rec.id: str(rec.seq) for rec in aln}

    # Agrupa os organismos pelo ambiente térmico (hipertermofilo, mesofilo, psicrófilo)
    grupos = {}
    for org in organismos:
        amb = ambientes[org]
        grupos.setdefault(amb, []).append(org)

    # Conta o total de posições no alinhamento e quantas são idênticas
    # em todos os organismos (conservadas) 
    total = aln.get_alignment_length()
    conservadas = sum(1 for i in range(total) if len(set(aln[:, i])) == 1)

    # Identifica posições onde cada grupo ambiental
    # tem aminoácidos completamente distintos dos outros grupos.
    # Essas são as posições candidatas a modificações impostas pelo ambiente.
    diagnosticas = []
    for i in range(total):
        # Para cada posição, coleta o conjunto de aminoácidos de cada grupo ambiental
        aa_por_grupo = {amb: set(seqs[org][i] for org in orgs)
                        for amb, orgs in grupos.items()}
        
        # Verifica se não há nenhum aminoácido em comum entre grupos diferentes
        # (interseção vazia entre todos os pares de grupos)
        todos_diferentes = all(
            not aa_por_grupo[a] & aa_por_grupo[b]
            for a in grupos for b in grupos if a != b
        )
        if todos_diferentes:
            diagnosticas.append(aa_por_grupo)

    # Conta a frequência de cada categoria química nas posições diagnósticas
    # separadamente por grupo ambiental
    contagem = {amb: Counter() for amb in grupos}
    for pos in diagnosticas:
        for amb, aas in pos.items():
            for aa in aas:
                contagem[amb][quimica.get(aa, '?')] += 1

    # Normaliza as contagens para proporções percentuais,
    # permitindo comparação entre grupos com números diferentes de organismos
    perfil = {}
    for amb, cont in contagem.items():
        total_amb = sum(cont.values())
        perfil[amb] = {cat: cont[cat]/total_amb*100 for cat in
                       ['apolar', 'polar', 'carregado+', 'carregado-']}

    # Armazena todos os resultados do gene para uso posterior nos gráficos
    resultados[gene] = {
        'total': total,
        'conservadas': conservadas,
        'diagnosticas': len(diagnosticas),
        'grupos': list(grupos.keys()),
        'perfil': perfil,
    }

    print(f"\n=== {gene} ===")
    print(f"  Total:        {total} aa")
    print(f"  Conservadas:  {conservadas} ({100*conservadas/total:.1f}%)")
    print(f"  Diagnósticas: {len(diagnosticas)}")
    print(f"  Grupos:       {list(grupos.keys())}")
```

### Análise do perfil químico nas posições diagnósticas

Com os alinhamentos prontos, a análise busca identificar se existe uma **assinatura
química distinta** associada a cada ambiente térmico nas sequências proteicas.

As posições do alinhamento em dois grupos:

- **Posições conservadas**: onde todos os organismos têm o mesmo aminoácido,
  independente do ambiente. Essas posições são essenciais para a função do gene
  e não variam porque qualquer mudança seria prejudicial ao organismo.

- **Posições diagnósticas**: onde cada grupo ambiental (hipertermófilo, mesófilo,
  psicrófilo) tem aminoácidos completamente distintos dos outros grupos. Essas são
  as posições candidatas a refletir adaptações impostas pelo ambiente, o organismo
  do frio tem um aminoácido, o do calor tem outro, e os dois nunca se misturam
  naquela posição.

Para cada posição diagnosticada, o aminoácido de cada organismo é classificado em uma
de quatro categorias químicas (apolar, polar, carregado+ ou carregado) e a
proporção de cada categoria é calculada separadamente por grupo ambiental.

Se o ambiente realmente imprime modificações sistemáticas no gene, espera-se que
cada grupo ambiental apresente um perfil químico distinto, por exemplo, organismos
do frio com maior proporção de aminoácidos polares e carregados+, e organismos do
calor com maior proporção de aminoácidos apolares, conforme descrito na literatura
para extremófilos (Verma et al., 2024).

O gráfico abaixo mostra esse perfil para os quatro genes analisados.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

cores = {'hipertermofilo': '#d62728', 'mesofilo': '#2ca02c', 'psicrófilo': '#1f77b4'}
categorias = ['apolar', 'polar', 'carregado+', 'carregado-']

for idx, (gene, dados) in enumerate(resultados.items()):
    ax = axes[idx]
    grupos_presentes = dados['grupos']
    x = np.arange(len(categorias))
    width = 0.8 / len(grupos_presentes)

    for i, amb in enumerate(grupos_presentes):
        vals = [dados['perfil'][amb].get(cat, 0) for cat in categorias]
        offset = (i - len(grupos_presentes)/2 + 0.5) * width
        bars = ax.bar(x + offset, vals, width,
                      label=amb, color=cores.get(amb, 'gray'))
        for b in bars:
            if b.get_height() > 0:
                ax.text(b.get_x() + b.get_width()/2, b.get_height() + 0.3,
                        f'{b.get_height():.1f}%', ha='center', va='bottom', fontsize=7)

    ax.set_title(f'{gene}\n({dados["diagnosticas"]} posições diagnósticas)', fontsize=11)
    ax.set_xticks(x)
    ax.set_xticklabels(categorias)
    ax.set_ylim(0, 55)
    ax.set_ylabel('Proporção (%)')
    ax.legend(fontsize=8)

plt.suptitle('Perfil químico nas posições diagnósticas por gene e ambiente térmico',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('resultados/grafico_todos_genes.png', dpi=150)
plt.show()
print("Gráfico salvo.")
```

### Comparação entre gyrA e reverse gyrase nos hipertermófilos

O mapeamento no KEGG revelou que organismos hipertermófilos lidam com a estabilidade
do DNA de duas formas distintas: alguns possuem o **gyrA** (DNA girase), outros possuem
a **reverse gyrase** e T. maritima possui os dois simultaneamente.

Ambas as enzimas atuam na topologia do DNA, mas de formas opostas:
- o **gyrA** relaxa o superenrolamento do DNA
- a **reverse gyrase** introduz superenrolamento positivo, estabilizando o DNA em
  temperaturas extremamente altas

Como as duas proteínas existem no mesmo ambiente térmico (> 80°C), a comparação entre
seus perfis químicos globais permite verificar se organismos do mesmo ambiente
convergem para composições de aminoácidos similares.

Diferentemente da análise anterior, que focou nas posições identificadas entre
ambientes, aqui seria o calculado do perfil químico global de cada proteína, ou seja,
a proporção de cada categoria química considerando todos os aminoácidos de todas as
sequências, não apenas as posições que diferem entre grupos. Isso permite comparar
a composição geral das duas proteínas diretamente.

Para o gyrA, apenas as sequências dos hipertermófilos (*T. ruber* e *T. maritima*)
são consideradas, para que a comparação com a reverse gyrase seja feita no mesmo
contexto ambiental.

```python
from Bio import AlignIO
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

quimica = {
    'A': 'apolar', 'V': 'apolar', 'I': 'apolar', 'L': 'apolar',
    'M': 'apolar', 'F': 'apolar', 'W': 'apolar', 'P': 'apolar',
    'G': 'apolar',
    'S': 'polar',  'T': 'polar',  'C': 'polar',  'Y': 'polar',
    'N': 'polar',  'Q': 'polar',
    'K': 'carregado+', 'R': 'carregado+', 'H': 'carregado+',
    'D': 'carregado-', 'E': 'carregado-',
    '-': 'gap'
}

def perfil_quimico(arquivo):
    aln = AlignIO.read(arquivo, "fasta")
    contagem = Counter()
    total_aa = 0
    for rec in aln:
        for aa in str(rec.seq):
            cat = quimica.get(aa)
            if cat and cat != 'gap':
                contagem[cat] += 1
                total_aa += 1
    return {cat: contagem[cat]/total_aa*100
            for cat in ['apolar', 'polar', 'carregado+', 'carregado-']}

# Calcular perfil de cada proteína
perfil_revgyrase = perfil_quimico("alinhamentos/revgyrase_alinhado.fasta")

# Perfil do gyrA só nos hipertermófilos
aln_gyra = AlignIO.read("alinhamentos/gyrA_alinhado.fasta", "fasta")
hipertermófilos_gyra = {"T_ruber", "T_maritima"}
contagem_gyra = Counter()
total_gyra = 0
for rec in aln_gyra:
    if rec.id in hipertermófilos_gyra:
        for aa in str(rec.seq):
            cat = quimica.get(aa)
            if cat and cat != 'gap':
                contagem_gyra[cat] += 1
                total_gyra += 1
perfil_gyra_quente = {cat: contagem_gyra[cat]/total_gyra*100
                      for cat in ['apolar', 'polar', 'carregado+', 'carregado-']}

# Mostrar valores
categorias = ['apolar', 'polar', 'carregado+', 'carregado-']
print(f"{'Categoria':<15} {'gyrA (quente)':>15} {'rev. gyrase':>15}")
print("-" * 48)
for cat in categorias:
    g = perfil_gyra_quente[cat]
    r = perfil_revgyrase[cat]
    print(f"{cat:<15} {g:>14.1f}% {r:>14.1f}%")

# Gráfico
x = np.arange(len(categorias))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2,
               [perfil_gyra_quente[c] for c in categorias],
               width, label='gyrA (hipertermófilos)', color='#d62728')
bars2 = ax.bar(x + width/2,
               [perfil_revgyrase[c] for c in categorias],
               width, label='reverse gyrase (hipertermófilos)', color='#ff7f0e')

for bars in [bars1, bars2]:
    for b in bars:
        ax.text(b.get_x() + b.get_width()/2, b.get_height() + 0.3,
                f'{b.get_height():.1f}%', ha='center', va='bottom', fontsize=9)

ax.set_title('Perfil químico global: gyrA vs reverse gyrase\n(ambos em organismos hipertermófilos)',
             fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(categorias)
ax.set_ylabel('Proporção de aminoácidos (%)')
ax.set_ylim(0, 60)
ax.legend()
plt.tight_layout()
plt.savefig('resultados/grafico_gyra_vs_revgyrase.png', dpi=150)
plt.show()
print("Gráfico salvo.")
```
